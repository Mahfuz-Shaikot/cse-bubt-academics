```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
```

```python
model = DiscreteBayesianNetwork([
    ("Rain", "Sprinkler"),
    ("Rain", "WetGrass"),
    ("Sprinkler", "WetGrass")
])

print(model.edges())
```

```python
cpd_rain = TabularCPD(
    variable="Rain",
    variable_card=2,
    values=[
        8[0.], [0.2]
    ]
)
```

```python
cpd_sprinkler = TabularCPD(
    variable="Sprinkler",
    variable_card=2,
    values=[
        [0.60, 0.99],
        [0.40, 0.01]
    ],
    evidence=["Rain"],
    evidence_card=[2]
)
```

```python
cpd_wetgrass = TabularCPD(
    variable="WetGrass",
    variable_card=2,
    values=[
        [0.99, 0.20, 0.10, 0.01],
        [0.01, 0.80, 0.90, 0.99]
    ],
    evidence=["Rain", "Sprinkler"],
    evidence_card=[2,2]
)
```

```python
model.add_cpds(
    cpd_rain,
    cpd_sprinkler,
    cpd_wetgrass
)
```

```python
print(model.check_model())
```

```python
infer=VariableElimination(model)
```

```python
print(infer.query(["Rain"]))
```

```python
print(infer.query(["WetGrass"]))
```

```python
print(
    infer.query(
        ["Rain"],
        evidence={"WetGrass":1}
    )
)
```

```python
print(
    infer.query(
        ["Sprinkler"],
        evidence={"WetGrass":1}
    )
)
```

```python
print(
    infer.query(
        ["WetGrass"],
        evidence={"Rain":1}
    )
)
```

```python
print(
    infer.query(
        ["WetGrass"],
        evidence={"Sprinkler":1}
    )
)
```

```python
from pgmpy.models import DiscreteBayesianNetwork

model = DiscreteBayesianNetwork([
    ("Burglary", "Alarm"),
    ("Earthquake", "Alarm"),
    ("Alarm", "JohnCalls"),
    ("Alarm", "MaryCalls")
])

print("Nodes:")
print(model.nodes())

print("\nEdges:")
print(model.edges())
```

```python
cpd_burglary = TabularCPD(
    variable="Burglary",
    variable_card=2,
    values=[
        [0.999],
        [0.001]
    ]
)
```

```python
cpd_earthquake = TabularCPD(
    variable="Earthquake",
    variable_card=2,
    values=[
        [0.998],
        [0.002]
    ]
)
```

```python
cpd_alarm = TabularCPD(
    variable="Alarm",
    variable_card=2,
    values=[
        [0.999, 0.71, 0.06, 0.05],
        [0.001, 0.29, 0.94, 0.95]
    ],
    evidence=["Burglary", "Earthquake"],
    evidence_card=[2, 2]
)
```

```python
cpd_john = TabularCPD(
    variable="JohnCalls",
    variable_card=2,
    values=[
        [0.95, 0.10],
        [0.05, 0.90]
    ],
    evidence=["Alarm"],
    evidence_card=[2]
)
```

```python
cpd_mary = TabularCPD(
    variable="MaryCalls",
    variable_card=2,
    values=[
        [0.99, 0.30],
        [0.01, 0.70]
    ],
    evidence=["Alarm"],
    evidence_card=[2]
)
```

```python
model.add_cpds(
    cpd_burglary,
    cpd_earthquake,
    cpd_alarm,
    cpd_john,
    cpd_mary
)
```

```python
print("Model is valid:", model.check_model())
```

```python
infer = VariableElimination(model)
```

```python
result = infer.query(variables=["JohnCalls"])

print(result)
```

```python
print(infer.query(variables=["MaryCalls"]))
```

```python
result = infer.query(
    variables=["JohnCalls"],
    evidence={
        "Alarm": 1,
        "Burglary": 0,
        "Earthquake": 0
    }
)

print(result)
```

```python
result = infer.query(
    variables=["MaryCalls"],
    evidence={
        "Alarm": 1,
        "Burglary": 0,
        "Earthquake": 0
    }
)

print(result)
```
